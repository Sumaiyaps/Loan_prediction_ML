```python
DROP COLUMNS
```

```python
data=data.drop("Unnamed: 2",axis=1)
```

```python
data=data.drop("Unnamed: 3",axis=1)
data=data.drop("Row Labels.1",axis=1)
data=data.drop("Sum of loan_amnt.1",axis=1)
data=data.drop("Unnamed: 6",axis=1)
data=data.drop("Unnamed: 7",axis=1)
data=data.drop("Row Labels.2",axis=1)
data=data.drop("Sum of loan_amnt.2",axis=1)
```

```python
data
```

```python
data['Sum of loan_amnt']=data['Sum of loan_amnt'].str.replace(',', '').astype(float)
```

```python
data['Row Labels'] = data['Row Labels'].astype(int)
```

```python
data.info()
```

```python
EXPLORATORY DATA ANALYSIS
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm',fmt='0.1f')
plt.show()
```

```python
FEATURE SELECTION
```

```python
x=data.drop('Sum of loan_amnt',axis=1)
y=data['Sum of loan_amnt']
```

```python
from sklearn.feature_selection import SelectKBest,f_regression
```

```python
bestfeatures=SelectKBest(score_func=f_regression,k=5)
fit=bestfeatures.fit(x,y)
```

```python
MODEL SPLITTING FOR TRAINING AND TESTING
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
MODEL SELECTION AND TRAINING

Linear Regression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print("r2_score:",r2_score(y_test,ypred))
```

```python
plt.scatter(y_test,ypred,color="skyblue")
plt.plot(y_test,y_test,color="red",linewidth=2)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("actual vs predicted")
plt.show()
```

```python
DECISION TREE
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
print("r2_score is:",r2_score(y_test,ypred))
```

```python
plt.scatter(y_test,ypred,color="red")
plt.plot(y_test,y_test,color="black",linewidth=2)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("actual vs predicted")
plt.show()
```

```python
RANDOM FOREST
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
print("r2 score:",r2_score(y_test,ypred))
```

```python
plt.scatter(y_test,ypred,color="green")
plt.plot(y_test,y_test,color="orange",linewidth=2)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("actual vs predicted")
plt.show()
```
